```python
import numpy as np

bert_pred = np.load('BERT-Predictions.npy')
hardv_pred = np.load('Hard Vote-Predictions.npy')
true_pred = np.load('test_labels.npy')
```

```python
import pandas as pd

test_df = pd.read_csv('/content/drive/MyDrive/CIS 5300 Term Project/Final_Data/test_data.csv')
```

```python
texts = test_df['text'].to_list()
```

```python
analysis_df = pd.DataFrame({
    "text": texts,
    "true_label": true_pred,
    "bert_pred": bert_pred,
    "hardv_pred": hardv_pred
})
```

```python
hardv_errors = analysis_df[
    analysis_df["hardv_pred"] != analysis_df["true_label"]
]
```

```python
hardv_fp = hardv_errors[
    (hardv_errors["hardv_pred"] == 1) &
    (hardv_errors["true_label"] == 0)
]
```

```python
hardv_fn = hardv_errors[
    (hardv_errors["hardv_pred"] == 0) &
    (hardv_errors["true_label"] == 1)
]
```

```python
with open("hard_vote_fp.txt", "w", encoding="utf-8") as f:
    for text in hardv_fp["text"]:
        f.write(text.strip())
        f.write("\n" + "-" * 80 + "\n")
```

```python
with open("hard_vote_fn.txt", "w", encoding="utf-8") as f:
    for text in hardv_fn["text"]:
        f.write(text.strip())
        f.write("\n" + "-" * 80 + "\n")
```
