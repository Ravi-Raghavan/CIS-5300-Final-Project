```python
import tarfile

# Path to your tar.gz file
file_path = "data_archive.tar.gz"

# Open the tar.gz file
with tarfile.open(file_path, "r:gz") as tar:
    # Extract all contents to the current directory
    tar.extractall(path="extracted_files")  # you can change this path
    # Optional: list all files in the archive
    print("Files in archive:")
    print(tar.getnames())
```
